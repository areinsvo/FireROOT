reco boundstate invariant mass by two lepton-jets

```python
from rootpy.io import root_open
import os
from FireROOT.Analysis.Utils import *
from rootpy.plotting.style import set_style
from rootpy.plotting import Canvas, Efficiency, Legend
from rootpy.plotting.shapes import Line
```

```python
fn = os.path.join(os.getenv('CMSSW_BASE'), 'src/FireROOT/Analysis/python/outputs/rootfiles/slimTrees/recoXX__TREE.root')
f = root_open(fn)
```

```python
f.keys()
```

```python
t=f.XX
sorted(t.branchnames)
```

```python
set_style(MyStyle())
canvas = Canvas(500, 400)
```

## $m_{\chi\bar{\chi}}$=500GeV

```python
CATEG0 = 'abs(s_MXX-500)<1e-6&abs(s_MA-0.25)<1e-6'
CATEG1 = 'abs(s_MXX-500)<1e-6&abs(s_MA-1.2)<1e-6'
CATEG2 = 'abs(s_MXX-500)<1e-6&abs(s_MA-5)<1e-6'
```

```python
mass_0 = t.Draw('xx_mass>>mass_0(100,250,750)', CATEG0, 'norm')
mass_1 = t.Draw('xx_mass>>mass_1(100,250,750)', CATEG1, 'norm')
mass_2 = t.Draw('xx_mass>>mass_2(100,250,750)', CATEG2, 'norm')

mass_0=mass_0.merge_bins([(0,1), (-2,-1)])
mass_1=mass_1.merge_bins([(0,1), (-2,-1)])
mass_2=mass_2.merge_bins([(0,1), (-2,-1)])
```

```python
canvas.clear()

hs = [mass_0, mass_1, mass_2]
for i,h in enumerate(hs):
    h.color=sigCOLORS[i]
    h.linewidth=2
    h.markersize=0.5

draw(hs[:], pad=canvas, xtitle='m_{#chi#bar{#chi}} [GeV]', ytitle='norm. counts/5GeV')

leg = Legend(3, margin=0.25, leftmargin=0.25,
             topmargin=0.03, entrysep=0.01, entryheight=0.03, textsize=14,)
leg.AddEntry(mass_0, 'm_{Z_{d}}=0.25GeV | #bar{x}: %.3f, #sigma: %.3f' % (mass_0.get_mean(), mass_0.get_std_dev()), 'LEP')
leg.AddEntry(mass_1, 'm_{Z_{d}}=1.2GeV | #bar{x}: %.3f, #sigma: %.3f' % (mass_1.get_mean(), mass_1.get_std_dev()), 'LEP')
leg.AddEntry(mass_2, 'm_{Z_{d}}=5GeV | #bar{x}: %.3f, #sigma: %.3f' % (mass_2.get_mean(), mass_2.get_std_dev()), 'LEP')
leg.Draw()

sampletag = LuminosityLabel('SIDM 2#mu2e m_{#chi#bar{#chi}}=500GeV, #bar{lxy}=150cm')
sampletag.draw()

canvas.draw()
```

```python
canvas.save_as('imgs/invm_xx500__2mu2e.pdf')
```
